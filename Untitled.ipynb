```python
def plt_classes(root_dir = "data"):
    fig, (axs) = plt.subplots(1,3,figsize=(15,5))
    splits = ["train","val","test"]
    for i,split in enumerate(splits):
        labels =  np.load(f"{root_dir}/{split}_labels.npy")
        x,y = np.unique(labels,return_counts=True)[0],np.unique(labels,return_counts=True)[1]
        axs[i].bar(x,y)
        axs[i].set_title(f"{split}")
    plt.show()
    
```
